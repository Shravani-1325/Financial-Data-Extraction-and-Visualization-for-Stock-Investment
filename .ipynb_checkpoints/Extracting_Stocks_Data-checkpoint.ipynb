```python
import yfinance as yf
import pandas as pd
import requests 
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

## Defining Graphing Function

The below function `make_graph` is designed to generate and display a dual-axis interactive visualization using Plotly.
It Compares a company's historical stock price with its historical revenue, allowing user to analyze trends and correlation between the two metrics overtime.

```python
def make_graph(stock_data, revenue_data, stock):
  fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing=0.3)

  # Filtering Data by Date
  stock_data_specific = stock_data[stock_data.Date <= "2021-06-14"]
  revenue_data_specific = revenue_data[revenue_data.Date <= "2021-04-30"]

  # Adding Stock Price Trace
  fig.add_trace(
      go.Scatter(
          x=pd.to_datetime(stock_data_specific.Date, format='%Y-%m-%d'),
          y=stock_data_specific.Close.astype("float"),
          name="Share Price"
      ),
      row=1, col=1
  )

  # Adding the Revenue Trace
  fig.add_trace(
      go.Scatter(
          x=pd.to_datetime(revenue_data_specific.Date, format='%Y-%m-%d'),
          y=revenue_data_specific.Revenue.astype("float"),
          name="Revenue"
      ),
      row=2, col=1
  )

  # Customizing the Axes
  fig.update_xaxes(title_text="Date", row=1, col=1)
  fig.update_xaxes(title_text="Date", row=2, col=1)
  fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
  fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)

  # Configure Layout
  fig.update_layout(
      showlegend=True,
      height=700,
      title=stock,
      xaxis_rangeslider_visible=True
  )

  # Displaying the Graphs
  fig.show()
```

## Extracting Tesla Stock using yfinance

```python
tesla = yf.Ticker("TSLA")
```

```python
# history of the stock
tesla_data = tesla.history(period="max")
```

```python
# Reseting the Index
tesla_data.reset_index(inplace=True)
```

```python
# Displaying the first 5 row
tesla_data.head()
```

## Webscraping to Extract Tesla Revenue Data

```python
url= "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text
```

```python
# Parsing the html data using beautiful_soup

soup = BeautifulSoup(html_data, "html.parser")
```

```python
# Total tables 

tables = soup.find_all("table")
len(tables)    
```

```python
# Searching Index of table Tesla Revenue

for index, table in enumerate(tables):
    if ("Tesla Quarterly Revenue" in str(table)):
        table_index = index
        break

if table_index is not None:
    print("Tesla Quarterly Revenue table Table found at index:", table_index)
else:
    print("Tesla Quarterly Revenue table not found.")
```

```python
# iterating through Rows in the table body

rows = tables[table_index].find_all("tr")
```

```python
#Extracting the data from columns

data =[]
for row in rows[1:]: # Skiping the header Row
    cols = row.find_all("td")
    date = cols[0].text.strip()
    revenue = cols[1].text.strip().replace("$", " ").replace(",","")
    data.append({"Date":date, "Revenue": revenue})

tesla_revenue = pd.DataFrame(data)
    
```

```python
# Removing the Null or Empty Strings

tesla_revenue.dropna(inplace=True)

tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
```

```python
# Displaying the last 5 row of the tesla_revenue

tesla_revenue.tail(5)
```
